```python
#image compression

import numpy as np
import matplotlib.pyplot as plt
from numpy import fft

data = np.loadtxt("http://www-personal.umich.edu/~mejn/cp/data/circular.txt",float) #upload data

ft = fft.fft2(data) #perform Fourier Transform 

plt.figure()
plt.imshow(data, plt.cm.gray)
plt.title('Original image')

plt.savefig("Original.jpeg",quality= 100) #size is 27,728 bytes 
```

```python
from numpy import fft
ft = fft.fft2(data)
F= np.log(abs(np.fft.fftshift(ft))+1)

plt.imshow(F, plt.cm.gray)
```

```python
rows = len(ft)

threshold = 5
for x in range(0, rows): 
    row = ft[x]
    columns = len(row)
    for y in range(0, columns):
        value = np.abs(ft[x][y])
        if value < threshold:
            ft[x][y] = 0 #values below threshold turn into zero 

ift= np.fft.ifft2(ft) #perform inverse Fourier Transform 
plt.imshow(ift.real,plt.cm.gray)

plt.savefig("Compressed.jpeg",quality= 100) #size is 25,862 bytes 
```

```python
#audio spectral analysis 
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fftpack import fft
import numpy as np

rate, data = wav.read('1102.wav') #uplod wav file 

xlist = list(range(0, len(data))) 
ylist = []
for x in xlist:
    ylist.append(data[x])

#print(len(data))
fft_out = fft(data) #perfrom FFT
%matplotlib inline
plt.plot(xlist, ylist)
#plt.plot(data, np.abs(fft_out))
plt.show() #first peaks is the first ring, and the second peak is the second ring in the audio file 
```
