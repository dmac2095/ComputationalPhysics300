```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=[15,10])

#define function
def f(x):
    return 3*np.cos(x+np.pi)

#parameters
a,b= 0,10
npoints= 100
order=11

#calculate ideal function
x= np.linspace(a,b,npoints) #start, end, number of points
y= f(x) 

#generate random noise
yy=f(x)+ np.random.rand(npoints)-.5

#get polyfit and evaluate polynomial using polyval
fit= np.polyfit(x,yy,order, full=True)
y_p= np.polyval(fit[0], x)

#calculate rsquared of function to determine if the function order is the right fit
ss_res= np.sum((y-y_p)**2) #sum of residuals squared 
ss_tot= np.sum((y-np.mean(y))**2) #total sum squared
r= 1-(ss_res/ss_tot) #rsquared 
               
print("R squared of the function is: "+str(r))

plt.plot(x,y ,label= "ideal function")
plt.plot(x, yy, 'o', label="random noise")
plt.plot(x,y_p,"--", label="polyfit"+str(order))

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.xlim([a,b])
plt.legend(fontsize=12)
plt.title('fitted coefficients' + str(fit[0]) + '\n residual: ' + str(fit[1]))
plt.show()

#looking at the rsquared value, order of 11 yields the best fit. Orders below 11 have lower rsquared values
#orders above 11 are not ideal because they capture the random noise
```
