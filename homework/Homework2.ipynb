```python
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
from math import log,exp

def factor(N):
    if N==0:
        return 1
    elif N==1:
        return 1
    else:
        return N*factor(N-1)
    
def multiplicity(qA, q, NA, NB):
    qB= q-qA
    
    a= log(factor(NA+qA-1))
    a2= log(factor(qA)*factor(NA-1))
    
    b =log(factor(NB+qB-1))
    b2=log(factor(qB)*factor(NB-1))
    
    aa=a-a2
    bb=b-b2
    
    return exp(aa+bb)

def Plot_Omega(q,NA,NB):

    qA= range(0,q+1)
    Omega= []
    
    for n in qA:
        Omega.append(multiplicity(n,q,NA,NB))
        
    plt.plot(qA, Omega, '.r')
    plt.title ("1. Two Interacting Einstein Solids")
    plt.xlabel('qA')
    plt.ylabel('$Multiplicity(qA)$')
    
Plot_Omega(100,300,200) #graph obtained shows that the most probable outcome is when 60% of 
                        #the total energy is in solid A
```

```python
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
from math import log,exp

def factor(N):
    total = 1
    if N<0:
        print('Error, N must be greater than 0')
    elif N>1: 
        for i in range(2,N+1):
            total *= i
    return total 

def multiplicity(qA, q, NA, NB):
    qB= q-qA
    
    a= log(factor(NA+qA-1))
    a2= log(factor(qA)*factor(NA-1))
    
    b =log(factor(NB+qB-1))
    b2=log(factor(qB)*factor(NB-1))
    
    aa=a-a2
    bb=b-b2
    
    return exp(aa+bb)

def Plot_Omega(q,NA,NB):

    qA= range(0,q+1)
    Omega= []
    
    for n in qA:
        Omega.append(multiplicity(n,q,NA,NB))
        
    plt.plot(qA, Omega, '^g')
    plt.title ("3. Two Interacting Einstein Solids")
    plt.xlabel('qA')
    plt.ylabel('$Multiplicity(qA)$')
    
Plot_Omega(10,30,20) #Comparing to plots 1&2 where N and q is large,  the peak in plot 4 is less narrow 
                     #This suggests that as N increases, the probability distribution gets closer to the value of the 
                         #most likely event 
```
